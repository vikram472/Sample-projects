```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\CODING\SHIASH\projects\Car price prediction\car.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum().sum()
```

```python
for i in data.select_dtypes(["float64"]).columns:
    if data[i].skew()>0.5:
        data[i]=data[i].fillna(data[i].median())
    else:
        data[i]=data[i].fillna(data[i].mean())
```

```python
data.isna().sum().sum()
```

```python
data=data.dropna()
```

```python
data.duplicated().sum()
```

```python
data=data.drop_duplicates(keep="last")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in data:
    if data[i].dtypes=="object":
        data[i]=le.fit_transform(data[i])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
d=scaler.fit_transform(data)
data=pd.DataFrame(d,columns=data.columns)
data
```

```python
x=data.drop(["selling_price"],axis=1)
y=data["selling_price"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
car_model=LinearRegression()
car_model.fit(x_train,y_train)
```

```python
pred=car_model.predict(x_test)
pred
```

```python
from sklearn.metrics import *
accuracy=r2_score(y_test,pred)
accuracy*100
```

```python
from sklearn.tree import DecisionTreeRegressor
de_house_model=DecisionTreeRegressor()
de_house_model.fit(x_train,y_train)
```

```python
pred1=de_house_model.predict(x_test)
pred1
```

```python
accuracy1=r2_score(y_test,pred1)
accuracy1*100
```
